```python
import numpy as np
import pandas as pd
import MakNN as mn
```

```python
#read dataset 
df = pd.read_csv('E:/Materials/Datasets/Placement_Data.csv',sep=',')
x = df.iloc[:,1:-1].values
y = df['status'].values

#create an object 
mn = mn.MakNN()

#normalize input
x = mn.normalize_input(x)

#split dataset to train val test
x_train, y_train, x_val, y_val, x_test, y_test = mn.split_data(x, y, train = 80, validation = 15, test = 5)
```

```python
#initialize your model
layers = [256, 512, 256, 1] #4 layers with their respected hidden units 
drop_out_param = [0.3,0.4,0.2,0] #dropouts at each layer, specify all 
activations = ['relu', 'relu', 'relu', 'sigmoid'] #activation functions at each layer 
input_shape = (x_train.shape[1], 8) #input dimension with the number of batchs [batch = 1 to use all the training examples]
batch_norm = [True, True, True, False]

mn.init_model(layers, activations, input_shape, drop_out_param = drop_out_param) #specify other parameters if required
```

```python
#train your model 
mn.train_model(x_train, y_train, x_val, y_val, epoch = 30, loss = 'logarithmic_loss', optimizer ='adam', learning_rate = 0.001, check_point = 'train_loss') #check if all the paramaeters work setting to different values 
```

```python
#make predictions [don't round the result if it is a regression problem]
ypred = np.round(mn.predict(x_test))  #never round this if you are working on a regression prblem
print("Predicted values", ypred)
#print("Actual values", mn.encode_y(y_test)) #never encode this if you are applying sigmoid 
print("Actual values", y_test)
```

```python
#load model and make prediction
mn.load_model()

ypred_saved = np.round(mn.predict(x_test))

print("from saved model", ypred_saved)
print("Actual values", y_test)
```
